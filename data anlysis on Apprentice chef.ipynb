```python
# importing necessary packages
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# opening data set for analysis
data=pd.read_excel("Apprentice_Chef_Dataset.xlsx")
df=pd.DataFrame(data)
```

ASSIGN THE DATA COLUMN TO ABBREVATED VARIABLES FOR EASY USE IN THE CODES

```python
# assigning data columns to variables for easy access
REV=data["REVENUE"]
TMO=data["TOTAL_MEALS_ORDERED"]
UMP=data["UNIQUE_MEALS_PURCH"]
CWCS=data["CONTACTS_W_CUSTOMER_SERVICE"]
PCV=data["PRODUCT_CATEGORIES_VIEWED"]
ATPS=data["AVG_TIME_PER_SITE_VISIT"]
MN=data["MOBILE_NUMBER"]
CBN=data["CANCELLATIONS_BEFORE_NOON"]
CAN=data["CANCELLATIONS_AFTER_NOON"]
TAP=data["TASTES_AND_PREFERENCES"]
PC_LOG=data["PC_LOGINS"]
M_LOG=data["MOBILE_LOGINS"]
WK=data["WEEKLY_PLAN"]
ED=data["EARLY_DELIVERIES"]
LD=data["LATE_DELIVERIES"]
PL=data["PACKAGE_LOCKER"]
RL=data["REFRIGERATED_LOCKER"]
APVT=data["AVG_PREP_VID_TIME"]
LOS=data["LARGEST_ORDER_SIZE"]
MCA=data["MASTER_CLASSES_ATTENDED"]
MMR=data["MEDIAN_MEAL_RATING"]
ACPV=data["AVG_CLICKS_PER_VISIT"]
TPV=data["TOTAL_PHOTOS_VIEWED"]
EMAIL=data["EMAIL"]

```

```python

z=data[["TOTAL_MEALS_ORDERED","UNIQUE_MEALS_PURCH","CONTACTS_W_CUSTOMER_SERVICE","PRODUCT_CATEGORIES_VIEWED","AVG_TIME_PER_SITE_VISIT","MOBILE_NUMBER","CANCELLATIONS_BEFORE_NOON","CANCELLATIONS_AFTER_NOON","TASTES_AND_PREFERENCES","MOBILE_LOGINS","PC_LOGINS","WEEKLY_PLAN","EARLY_DELIVERIES","LATE_DELIVERIES","PACKAGE_LOCKER","REFRIGERATED_LOCKER","AVG_PREP_VID_TIME","LARGEST_ORDER_SIZE","MASTER_CLASSES_ATTENDED","MEDIAN_MEAL_RATING","AVG_CLICKS_PER_VISIT","TOTAL_PHOTOS_VIEWED"]]
y=REV
```

FIT THE DATA IN A MULTIPLE REGRESSION MODEL

```python
# fit data into ols model
x=sm.add_constant(z)
model=sm.OLS(y,x)
result=model.fit()
result.summary() # ols results

```
